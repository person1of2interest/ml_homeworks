# 1. Решающие деревья. Визуализация.

В этой части мы рассмотрим два простых двумерных датасета сделанных с помощью `make_moons`, `make_circles` и посмотрим как ведет себя разделяющая поверхность в зависимости от различных гиперпараметров.

```python
from sklearn.datasets import make_moons, make_circles, make_classification
datasets = [
    make_circles(noise=0.2, factor=0.5, random_state=42),
    make_moons(noise=0.2, random_state=42),
    make_classification(n_classes=3, n_clusters_per_class=1, n_features=2, class_sep=.8, random_state=3,
                        n_redundant=0.)
]
```

```python
palette = sns.color_palette(n_colors=3)
cmap = ListedColormap(palette)
```

```python
plt.figure(figsize=(15, 4))
for i, (x, y) in enumerate(datasets):
    plt.subplot(1, 3, i + 1)
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap, alpha=.8)
```

__Задание 1. (1 балл)__

Для каждого датасета обучите решающее дерево с параметрами по умолчанию, предварительно разбив выборку на обучающую и тестовую. Постройте разделящие поверхности (для этого воспользуйтесь функцией `plot_surface`, пример ниже). Посчитайте accuracy на обучающей и тестовой выборках. Сильно ли деревья переобучились?

```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

```python
# Пример:
from sklearn.linear_model import LinearRegression
X, y = datasets[2]
lr  = LinearRegression().fit(X, y)
plot_surface(lr, X, y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for i, (X, y) in enumerate(datasets):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
    
    dt = DecisionTreeClassifier(random_state=1000)
    dt.fit(X_train, y_train)
    
    axs[i] = plt.subplot(1, 3, i+1)
    plot_surface(dt, X_test, y_test)
    axs[i].set_title(f'accuracy-train: {accuracy_score(y_train, dt.predict(X_train)):.2f},\n' +
                     f'accuracy-test: {accuracy_score(y_test, dt.predict(X_test)):.2f}')
```

__Ответ:__ Деревья переобучились сильно (модели полностью предсказывают обучающие выборки), но, тем не менее, вторая модель даёт хорошую точность предсказаний на тестовой выборке.

Попробуйте перебрать несколько параметров для регуляризации (напр. `max_depth`, `min_samples_leaf`). Для каждого набора гиперпараметров постройте разделяющую поверхность, выведите обучающую и тестовую ошибки. Можно делать кросс-валидацию или просто разбиение на трейн и тест, главное делайте каждый раз одинаковое разбиение, чтобы можно было корректно сравнивать (помните же, что итоговое дерево сильно зависит от небольшого изменения обучающей выборки?). Проследите как меняется разделяющая поверхность и обобщающая способность. Почему так происходит, одинаково ли изменение для разных датасетов?

**Для первого датасета:**

```python
X, y = datasets[0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

Вывожу *accuracy* для удобства (ошибка на выборке получается вычитанием *accuracy* из единицы).

```python
_, axs = plt.subplots(3, 3, figsize=(15, 15))

for i, max_depth in enumerate([2, 4, None]):
    for j, min_samples_leaf in enumerate([15, 5, 1]):
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        dt.fit(X_train, y_train)
        
        axs[i][j] = plt.subplot(3, 3, 3*i+j+1)
        plot_surface(dt, X_test, y_test)
        axs[i][j].set_title('max_depth=' + str(max_depth) + 
                            ' | min_samples_leaf=' + str(min_samples_leaf) + '\n' +
                            f'accuracy-train: {accuracy_score(y_train, dt.predict(X_train)):.2f} | ' +
                            f'accuracy-test: {accuracy_score(y_test, dt.predict(X_test)):.2f}')
```

**Для второго датасета:**

```python
X, y = datasets[1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

```python
_, axs = plt.subplots(3, 3, figsize=(15, 15))

for i, max_depth in enumerate([2, 4, None]):
    for j, min_samples_leaf in enumerate([15, 5, 1]):
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        dt.fit(X_train, y_train)
        
        axs[i][j] = plt.subplot(3, 3, 3*i+j+1)
        plot_surface(dt, X_test, y_test)
        axs[i][j].set_title('max_depth=' + str(max_depth) + 
                            ' | min_samples_leaf=' + str(min_samples_leaf) + '\n' +
                            f'accuracy-train: {accuracy_score(y_train, dt.predict(X_train)):.2f} | ' +
                            f'accuracy-test: {accuracy_score(y_test, dt.predict(X_test)):.2f}')
```

**Для третьего датасета:**

```python
X, y = datasets[2]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

```python
_, axs = plt.subplots(3, 3, figsize=(15, 15))

for i, max_depth in enumerate([2, 4, None]):
    for j, min_samples_leaf in enumerate([15, 5, 1]):
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        dt.fit(X_train, y_train)
        
        axs[i][j] = plt.subplot(3, 3, 3*i+j+1)
        plot_surface(dt, X_test, y_test)
        axs[i][j].set_title('max_depth=' + str(max_depth) + 
                            ' | min_samples_leaf=' + str(min_samples_leaf) + '\n' +
                            f'accuracy-train: {accuracy_score(y_train, dt.predict(X_train)):.2f} | ' +
                            f'accuracy-test: {accuracy_score(y_test, dt.predict(X_test)):.2f}')
```

__Ответ:__ 
- Для первого датасета оптимальным из трёх значений `min_samples_leaf` (то есть числа элементов в вершине, при котором вершина считается листом) является 5, а глубина >=4.
- Для второго датасета уменьшение `min_samples_leaf` ведёт к увеличению точности как на тренировочной выборке, так и на тестовой. Лучший результат, как ни странно, достигается, когда модель просто запоминает обучающую выборку.
- Для третьего датасета при `min_samples_leaf`=1 наблюдается явное переобучение, а остальные случаи дают одинаковые разделяющие поверхности и значения точности.
- Таким образом, изменения для разных датасетов разные, но в целом верно, что увеличение максимальной глубины и уменьшение `min_samples_leaf` до некоторого момента улучшают обобщающую способность модели.

Проанализируйте результаты эксперимента. 
Одинаково ли для разных наборов данных ранжируются алгоритмы? 
Порассуждайте, почему так происходит. 

Обратите внимание на значение признаков в разных наборах данных. 
Присутствует ли в результатах какая-то компонента случайности? 
Можно ли повлиять на нее и улушить работу алгоритмов?

**Ответ:**
- Результаты ранжируются неодинаково. Например, для датасета 'tic-tac-toe' лучший результат из трёх вариантов решающего дерева даёт дерево, которое обучается на закодированных при помощи One-Hot Encoding признаков, а для 'mushrooms' лучшим вариантом оказывается модель, обученная на Ordinal Encoding категориальных признаках. Понятно, почему она почти на всех датасетах даёт лидирующий результат. При этом известно, что дерево может сильно переобучиться, и модели (а следовательно, и результаты) отличаются даже для разных подвыборок, не говоря уже о разных датасетах, отсюда разное поведение.
- В датасете 'nursery' значения признака health часто совпадают с целевой переменной, поэтому, возможно, объединение нескольких значений в одно аналогично тому, как это делается для целевой переменной, немного повысило бы качество результата и скорость работы алгоритма. В целом для всех датасетов можно попробовать использовать стратифицированную выборку относительно целевой переменной.
